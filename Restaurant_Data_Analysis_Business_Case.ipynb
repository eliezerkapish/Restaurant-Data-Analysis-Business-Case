Data Analysis of Restaurant Variables to Identify the Revenue Impact of Different Factors.

The analysis includes the following variables: Marketing Budget, Social Media Followers, Seating Capacity, Average Meal Price, Parking Availability, Weekend Reservations, Weekday Reservations, Ambience Score, Service Quality Score, Chef Experience, and Years.

The purporse of this analysis is to identify the higher correlation variable to generate revenue.

# IMPORT LIBRARIES AND DATAESTS


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

# READ DATASETS

```python
dataset_restaurant_csv = pd.read_csv('/content/drive/MyDrive/Restaurante/Dataset_Restaurant.csv')
dataset_restaurant_xlsx = pd.read_excel('/content/drive/MyDrive/Restaurante/Dataset_Restaurant.xlsx')
restaurant_location_csv = pd.read_csv('/content/drive/MyDrive/Restaurante/Restaurant_Location.csv')
restaurant_location_xlsx = pd.read_excel('/content/drive/MyDrive/Restaurante/Restaurant_Location.xlsx')
```

# CONCAT DATASETS CSV AND XLSX TO VERIFY ADITIONAL INFORMATION

```python
dataset_restaurant = pd.concat([dataset_restaurant_csv, dataset_restaurant_xlsx], ignore_index=True)
restaurant_location = pd.concat([restaurant_location_csv, restaurant_location_xlsx], ignore_index=True)
```

#CONVERT TO NUMERIC VALUES

```python
# Convert the 'Parking Availability' to numeric value
dataset_restaurant['Parking Availability'] = dataset_restaurant['Parking Availability'].map({'Yes': 1, 'No': 0})

# Select the variables of interest for the correlation analysis
variables_of_interest = [
    'Revenue',
    'Marketing Budget',
    'Social Media Followers',
    'Seating Capacity',
    'Average Meal Price',
    'Parking Availability',
    'Weekend Reservations',
    'Weekday Reservations',
    'Ambience Score',
    'Service Quality Score',
    'Chef Experience Years'
]


```

# CALCULATE CORRELATION MATRIX


```python
correlation_matrix = dataset_restaurant[variables_of_interest].corr()
```

# CORRELATION WITH REVENUE

```python
correlation_with_revenue = correlation_matrix['Revenue'].sort_values(ascending=False)
correlation_with_revenue
```

#PLOT CORRELATION MATRIX

```python
# Calculation Matrix
correlation_matrix = dataset_restaurant[variables_of_interest].corr()

# Plot Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Between Variables')
plt.show()
```

#CONCLUSION

The correlation of 0.68 with revenue indicates a strong positive relationship between the average meal price and the restaurant's revenue. This suggest that restaurants that charge more per meal tend to generate higher revenue. The seating capacity also shows a strong positive relationship, with a correlation of 0.67 indicating that restaurants with greater seating capacity tend to genereate higher revenue.
